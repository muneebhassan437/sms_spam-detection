# Loading the dataset

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset
import evaluate

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python
# Load the spam Dataset
raw_datasets = load_dataset("ucirvine/sms_spam")
print(raw_datasets)

sms_dataset = raw_datasets["train"].train_test_split(test_size=0.2, seed=42)

train_dataset = sms_dataset["train"]
test_dataset  = sms_dataset["test"]

print("Train size:", len(train_dataset))
print("Test size:", len(test_dataset))
```

# Bert model

```python
# Load a Pretrained BERT and Tokenizer
pretrained_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(pretrained_name)

model = BertForSequenceClassification.from_pretrained(
    pretrained_name,
    num_labels=2  # sms: spam(1) or not spam (0)
)
model.to(device)
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["sms"],  # Change this to your text column name
        padding="max_length",
        truncation=True,
        max_length=256
    )

train_dataset = train_dataset.map(tokenize_function, batched=True)
test_dataset  = test_dataset.map(tokenize_function, batched=True)
```

```python
# We no longer need the original "text" column
train_dataset = train_dataset.remove_columns(["sms"])
test_dataset  = test_dataset.remove_columns(["sms"])
```

```python
# Define Accuracy Metric
accuracy_metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = torch.from_numpy(logits).argmax(dim=-1)
    return accuracy_metric.compute(predictions=preds, references=labels)
```

```python
# Set Up Training Trainer and Fine-Tuning
training_args = TrainingArguments(
    output_dir="./bert-ucirvine/sms_spam",
    evaluation_strategy="no",  # Disable evaluation
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_steps=100,
    load_best_model_at_end=False,  # No best model selection without eval
    fp16=True if device == "cuda" else False,
    report_to=["none"],
    logging_dir="./logs",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
test_results = trainer.evaluate(test_dataset)
print("Test set results:", test_results)
```

```python
# Load the IMDB Dataset
raw_datasets = load_dataset("ucirvine/sms_spam")
print(raw_datasets)

sms_dataset = raw_datasets["train"].train_test_split(test_size=0.2, seed=42)

train_dataset = sms_dataset["train"]
test_dataset  = sms_dataset["test"]

```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch
import pandas as pd
from sklearn.metrics import accuracy_score

# Load model and tokenizer
model_name = "typeform/distilbert-base-uncased-mnli"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.eval()  # Set model to evaluation mode

def predict_spam(text):
    # Prepare the prompt
    prompt = f"""
    Classify the following text into one of these categories: spam, ham.
    Text: "{text}"
    Answer:
    """

    # Tokenize and predict
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)

    # Get probabilities
    probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)

    # Return prediction (1 for spam, 0 for ham)
    return 1 if probabilities[0][1] > probabilities[0][0] else 0



my_label = test_dataset['label']
my_sms = test_dataset['sms']

predicted_label = []
for sms in my_sms:
  predicted_label.append(predict_spam(sms))
print(predicted_label)
```

```python
accuracy = accuracy_score(predicted_label, my_label)
print(f"Accuracy: {accuracy:.4f}")
```

```python
from datasets import load_dataset
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
dataset = load_dataset("sms_spam")

# Extract texts and labels
texts = [example['sms'] for example in dataset['train']]
labels = [1 if example['label'] == 1 else 0 for example in dataset['train']]

# Check distribution
print(f"Spam: {sum(labels)}, Ham: {len(labels) - sum(labels)}")

# Convert text data into BoW features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Stratified split to maintain class balance
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42, stratify=labels)

# Train logistic regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
# Assign label 0 to every test datapoint
y_pred_dummy = [0] * len(y_test)

# Calculate accuracy
dummy_accuracy = accuracy_score(y_test, y_pred_dummy)
print(f"Dummy Model Accuracy: {dummy_accuracy:.2f}")
```
